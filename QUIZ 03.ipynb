```python
# 1. list 내포 - 피타고라스 정리
# list 내포를 사용하여 아래 조건을 만족하는 결과 list를 만드시오 (단, 한줄 코딩).
#  1) 직각 삼각형의 각 변의 길이는 a, b, c 이며 a^2 + b^2 = c^2 이다.
#  2) a와 b의 범위는 각각 1~100, c의 범위는 1~1000 일 때, 가능한 직각삼각형의 변의 길이 집합 (a, b, c)를 원소로 갖는 vector를 구하라 (단, a <= b < c).

result = [(a,b,c)  if a*a + b*b == c*c else None for a in range(1,101) for b in range(a,101) for c in range(b+1,1001)]
result = list(filter(None, result))
print(result)
```

```python
# 2. list 내포 - 소수 만들기
# list 내포를 사용하여 아래 조건을 만족하는 결과 list를 만드시오 (단, 한줄 코딩).
#  1) 결과 list의 범위는 정수 2~1000 이다.
#  2) 결과 list는 소수로 이루어진 vector이다.

result = list(filter(
    None, [
        num[-1] if all ([False if num[-1] != k and num[-1]%k == 0 else num[-1] for k in num]) else None
        for num in [[j for j in range(2, i+1)] for i in range(2, 1001)]]
))
result = list(filter(None, result))
print(result)
```
